4.What output does this code produce? Why?

&gt;&gt;&gt; X = &#39;iNeuron&#39;                                                               
&gt;&gt;&gt; def func():                                                                  
global X                                                                                
X = &#39;NI&#39;                                                                                
&gt;&gt;&gt; func()                                                                                
&gt;&gt;&gt; print(X)                                          

Ans : The output of the code is NI. the global keyword allows a varible to be accessible in the current scope. since we are using global keyword inside the funtion func it directly access the variable in X in global scope. and changes its value to NI . hence the output of the code is NI

```python
X = 'iNeuron'
def func():
    global X
    X = 'NI'
func()
print(X)
```
